```python
df['comment'] = ""
```

```python
for i in range(len(df)):
  condition = df.iloc[i].to_dict()
  #print(condition)
  #print()

  # 질문 작성하기
  query = f"""국민은행의 예금 상품, 투자 및 적립 상품 중에서 나이 {condition['age']}세, 직업 {condition['job']}, {condition['marital']},
  {condition['grade']} 등급, 신용불량 {condition['default']}, 예금잔고 {condition['balance']}달러, 주택담보대출 {condition['housing']},
  대출 {condition['loan']}, IRP {condition['IRP']}인 손님을 위한 상품 추천 멘트를 실제로 말하듯이 작성해줘"""

  # print(query)
  # print()

  # 메시지 설정하기
  messages = [
          {"role": "system", "content": "You are a helpful assistant with korean."},
          {"role": "user", "content": query}
  ]

  # ChatGPT API 호출하기
  response = openai.ChatCompletion.create(
      model=model,
      messages=messages
  )
  answer = response['choices'][0]['message']['content'].replace('\n','')
  df['comment'][i] = answer

  time.sleep(0.5)
  # print(answer)
  # print("-"*50)
```

```python
#OpenAI API의 시간적 한계로 인해 timeout 현상 발생 -> 총 9천여개 중에서 54개만 comment 생성 완료
df_rcmd = df[df['comment']!=""]
df_rcmd
```
